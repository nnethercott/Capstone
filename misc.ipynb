```python
#goal, 3d plotting tumour contours, lets first try to make a sphere 

def showslices(slices, figsize=(10,10), vert = 0.1):

    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111, projection='3d')

    x = np.linspace(0, slices[0].shape[1]-1, slices[0].shape[1])
    y = x
    X,Y = np.meshgrid(x,y)
    
    #now some heights 
    heights = [vert*i for i in range(len(slices))]
    
    ax.set_zlim3d(0, heights[-1])
    #now plot
    for i, s in enumerate(slices):
        ax.contour(X,Y, s+heights[i]*np.ones(s.shape), [heights[i]])
    
```
